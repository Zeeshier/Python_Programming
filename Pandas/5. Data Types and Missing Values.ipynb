```python
import pandas as pd
data = pd.read_csv('Cuisine_rating.csv')
data.head()
```

# Dtypes

```python
# Dtype : to grab the type of a specific column

data['Overall Rating'].dtype
```

```python
# dtypes : returns the dtype of every column in the DataFrame

data.dtypes
```

```python
# astype() : to convert a column of one type into another

data['Food Rating'].astype('float64')
```

```python
# A DataFrame or Series index has its own dtype

data.index.dtype
```

# Missing data


```python
# isnull() : specific to missing data

data[pd.isnull(data.Location)]
```

```python
# fillna() : for mitigating such data

data.fillna('Unknown')
```

```python
# replace() : non-null value that we would like to replace

data['Location'].replace("Upper East Side,NY", "Lahore")
```
