# Groupwise analysis

```python
import pandas as pd
data = pd.read_csv('Cuisine_rating.csv')
data.head()
```

```python
# groupby()

data.groupby('Food Rating')['Food Rating'].count()
```

```python
data.groupby('Food Rating')['Budget'].min()
```

```python
data.groupby('Cuisines').apply(lambda df: df['Overall Rating'].iloc[0])
```

```python
data.groupby(['Area code', 'Location']).apply(lambda df: df.loc[df.Budget.idxmax()])
```

```python
# agg():  run a bunch of different functions on your DataFrame simultaneously

data.groupby(['Location']).Budget.agg([len, min, max])
```

# Multi-indexes


```python
group = data.groupby(['Area code', 'Location']).Cuisines.agg([len])
group
```

```python
r = group.index
type(r)
```

```python
#  reset_index()

group.reset_index()
```

# Sorting

```python
# sort by len

group = group.reset_index()
group.sort_values(by= 'len')
```

```python
# descending sort

group.sort_values(by= 'len', ascending=False)
```

```python
# To sort by index values

group.sort_index()
```

```python
group.sort_values(by=['Location', 'len'])
```
