# Groupwise analysis

```python
import pandas as pd
data = pd.read_csv('Cuisine_rating.csv')
data.head()
```

```python
# groupby()

data.groupby('Food Rating')['Food Rating'].count()
```

```python
data.groupby('Food Rating')['Budget'].min()
```

```python
data.groupby('Cuisines').apply(lambda df: df['Overall Rating'].iloc[0])
```

```python
data.groupby(['Area code', 'Location']).apply(lambda df: df.loc[df.Budget.idxmax()])
```

```python
# agg():  run a bunch of different functions on your DataFrame simultaneously

data.groupby(['Location']).Budget.agg([len, min, max])
```
